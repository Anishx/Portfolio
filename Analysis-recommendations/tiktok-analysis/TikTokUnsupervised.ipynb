#### **Engagement** 
- Less than 1% = low engagement rate. 
- Between 1% and 3.5% = average/good engagement rate. 
- Between 3.5% and 18% = high engagement rate. 
- Above 18% = very high engagement rate.

We're unable to see if an account belongs to a specific gender without more accurate data. 

```python
def classify_engagement_rate(rate):
    rate_dict = {
        'low engagement rate': lambda x: x < 1,
        'average/good engagement rate': lambda x: 1 <= x <= 3.5,
        'good engagement rate': lambda x: 3.5 < x <= 18,
        'very high engagement rate': lambda x: x > 18
    }
    for key, func in rate_dict.items():
        if func(rate):
            return key

df['engagement_rate_category'] = df['awg_engagement_rate'].apply(classify_engagement_rate)
```

#### **Clustering**
Considering the numerical features of the dataset which can contribute to segmenting the users

```python
features = ['followers','comment_engagement_rate', 'like_engagement_rate', 'likes', 'videos_count','following']
X = df[features]
```

```python
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
```

##### **Applying K-Means considering all variables**

```python
kmeans_model = KMeans(random_state=0, n_init=10)
visualizer = KElbowVisualizer(kmeans_model, k=(2,10), timings=True)

visualizer.fit(X_scaled)  
visualizer.show() 

optimal_clusters = visualizer.elbow_value_
kmeans_optimal = KMeans(n_clusters=optimal_clusters,n_init=10, random_state=0).fit(X_scaled)
cluster_labels = kmeans_optimal.labels_
centroids = kmeans_optimal.cluster_centers_  

print('Optimal clusters: ',optimal_clusters)
```

```python
optimal_k = 5
kmeans_final = KMeans(n_clusters=optimal_k, random_state=0, n_init=10)
kmeans_final.fit(X_scaled)
df['cluster'] = kmeans_final.labels_
```

```python
cluster_summary = df.groupby('cluster')[features].mean()

print("General User Segments Summary:")
cluster_summary
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X.toarray())
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
finalDf = pd.concat([principalDf, df[['cluster']]], axis = 1)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
colors = ['r', 'g', 'b', 'y', 'c', 'm']
for cluster, color in zip(df['cluster'].unique(), colors):
    indicesToKeep = finalDf['cluster'] == cluster
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(df['cluster'].unique())
ax.grid()
```
